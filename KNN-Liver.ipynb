```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn

from google.colab import files
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_iris
from sklearn.linier_model import LogisticRegression

uploaded = files.upload()

datas = pd.read_csv("csv_KNN_Liver_CSU_1.csv")
datas.head()

def devided_data(datas, name_colums=['ALB','ALP','ALT','AST','BIL','PROT','GGT']):
  ciri = datas[name_colums]
  label = datas[['Diagnosa']]
  return ciri,label

ciri,label = devided_data(datas)

Ciri_train, Ciri_test, Label_train, Label_test = train_test_split(ciri, label, test_size = 0.10, random_state = 0) #par

classifier = KNeighborsClassifier(n_neighbors = 1, metric = 'minkowski', p = 2)
classifier.fit(Ciri_train, Label_train.values.ravel())

Label_pred = classifier.predict(Ciri_test)

cm = confusion_matrix(Label_test, Label_pred)
ac = accuracy_score(Label_test, Label_pred)

```
